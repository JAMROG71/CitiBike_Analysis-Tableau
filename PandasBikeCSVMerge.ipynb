```python
import pandas as pd

files = ['201806citibiketripdata.csv', '201807citibiketripdata.csv', '201808citibiketripdata.csv', '201809citibiketripdata.csv']

df = pd.DataFrame()
for filename in files:
    temp_df = pd.read_csv(filename)
    df = pd.concat([df, temp_df], ignore_index=True)

df.to_csv('summer_2018.csv', index=False)
```
